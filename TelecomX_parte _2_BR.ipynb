#ðŸ“Œ ExtracÃ£o

```python
import json


#Load the JSON data

with open('TelecomX_Data.json','r') as f:
    data = json.load(f)

print(data)
```

#ðŸ”§ TransformaÃ§Ã£o

```python
import pandas as pd

df = pd.DataFrame(data)

print(df)
```

```python
# Separa apenas as colunas Customer ID, Churn e Account

df_churn= df[['customerID', 'Churn', 'account']]

print(df_churn)
```

```python
# Considerando apenas as linhas de Churn que contÃ©m Yes

df_churnY = df_churn[df_churn['Churn'] == 'Yes']

print(df_churnY)
```

#ðŸ“Š Carga e anÃ¡lise

```python
# Analisando o tipo de contrato - month-to-month

df_churnMM = df_churnY[df_churnY['account'].apply(lambda x: x['Contract'] == 'Month-to-month')]

print(df_churnMM)
```

```python
# Analisando o tipo de contrato - one year

df_churnOY = df_churnY[df_churnY['account'].apply(lambda x: x['Contract'] == 'One year')]

print(df_churnOY)
```

```python
# Analisando o tipo de contrato - two year

df_churnTY = df_churnY[df_churnY['account'].apply(lambda x: x['Contract'] == 'Two year')]

print(df_churnTY)
```

```python
# Contando a quantidade de pessoas com contrato - Two year

count_churnTY = len(df_churnTY)

print(count_churnTY)
```

```python
# Contando a quantidade de pessoas evasivas

count_churnY = len(df_churnY)

print(count_churnY)
```

```python
# Analisando a porcentagem dos 3 tipos de contrato

percentage_churnMM = (len(df_churnMM) / count_churnY) * 100
percentage_churnOY = (len(df_churnOY) / count_churnY) * 100
percentage_churnTY = (len(df_churnTY) / count_churnY) * 100

print(percentage_churnMM)
print(percentage_churnOY)
print(percentage_churnTY)
```

```python
# Arredondando as mÃ©dias adquiridas

percentage_churnMM = round(percentage_churnMM, 2)
percentage_churnOY = round(percentage_churnOY, 2)
percentage_churnTY = round(percentage_churnTY, 2)

print(f'A porcentagem do plano mensal Ã©: {percentage_churnMM} %')
print(f'A porcentagem do plano anual Ã©: {percentage_churnOY} %')
print(f'A porcentagem do plano biÃªnio Ã©: {percentage_churnTY} %')
```

```python
import matplotlib.pyplot as plt

# Plotting the percentage pie chart

labels = ['Month-to-month', 'One year', 'Two year']
sizes = [percentage_churnMM, percentage_churnOY, percentage_churnTY]
colors = ['#ff9999','#66b3ff','#99ff99']
explode = (0.1, 0, 0)  # explode 1st slice

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage of Churned Customers by Contract Type')
plt.show()
```
